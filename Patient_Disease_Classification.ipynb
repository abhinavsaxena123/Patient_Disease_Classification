```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words("english")
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer

porter = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
from bs4 import BeautifulSoup
import re
```

```python
def review_to_words(raw_review):
    review_text = BeautifulSoup(raw_review, 'html.parser').get_text()
    letters_only = re.sub('[^a-zA-Z]', ' ', review_text)
    words = letters_only.lower().split()
    #removing stopwords
    meaningful_words = [word for word in words if not word is stop]
    #lemmatization
    lemmatize_words = [lemmatizer.lemmatize(word) for word in meaningful_words]
    return (' '.join(lemmatize_words))
```

```python
X['review_clean'] = X['review'].apply(review_to_words)
```

```python
X.head()
```

### Creating Features and Target Variable

```python
X_feat = X['review_clean']
y = X['condition']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_feat,y,stratify=y,test_size=0.2,random_state=0)
```

```python
def plot_confusion_matrix(cm,classes,
                          title='Confusion Matrix',
                          cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

     # Normalize the confusion matrix if desired
    fmt = '.2f'
    thresh = cm.max() / 2.
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

### Bag of Words (Count Vectorizer)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
count_train = cv.fit_transform(X_train)
count_test = cv.transform(X_test)
```

### ML Model: Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

#Train the model
model = MultinomialNB()
model.fit(count_train,y_train)

# Make predictions
pred = model.predict(count_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy:.3f}")

class_names=['Birth Control','Depression','High Blood Pressure','Obesity','Bacterial Infection']
cm = confusion_matrix(y_test, pred, labels=class_names)
plot_confusion_matrix(cm, classes=class_names)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

## TF-IDF 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

v = TfidfVectorizer(stop_words='english',max_df=0.8)
tfidf_train_2 = v.fit_transform(X_train)
tfidf_test_2 = v.transform(X_test)
```

## Naive Bayes

```python
mnb_tf = MultinomialNB()
mnb_tf.fit(tfidf_train_2,y_train)
tf_pred = mnb_tf.predict(tfidf_test_2)

# Calculate accuracy
accuracy = accuracy_score(y_test, tf_pred)
print(f"Accuracy: {accuracy:.3f}")

class_names=['Birth Control','Depression','High Blood Pressure','Obesity','Bacterial Infection']
cm = confusion_matrix(y_test, tf_pred, labels=class_names)
plot_confusion_matrix(cm, classes=class_names)
```

```python
print(classification_report(y_test, tf_pred))
```

### TFIDF AND Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer(stop_words='english', max_df=0.8)),        #using the ngram_range parameter 
     ('Random Forest', RandomForestClassifier())         
])

# 2. Fit the pipeline with raw text data (X_train) and labels (y_train)
clf.fit(X_train, y_train)

#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.3f}")

class_names=['Birth Control','Depression','High Blood Pressure','Obesity','Bacterial Infection']
cm = confusion_matrix(y_test, y_pred, labels=class_names)
plot_confusion_matrix(cm, classes=class_names)
```

```python
print(classification_report(y_test, y_pred))
```

### CountVectorizer Bigrams

```python
#1. create a pipeline object
clf_bi = Pipeline([
     ('vectorizer_bow', CountVectorizer( ngram_range = (1, 2), max_df=0.8)),  #using the ngram_range parameter 
     ('Multi_NB', MultinomialNB())         
])

#2. fit with X_train and y_train
clf_bi.fit(X_train, y_train)

#3. get the predictions for X_test and store it in y_pred
y_pred_bi = clf_bi.predict(X_test)

class_names=['Birth Control','Depression','High Blood Pressure','Obesity','Bacterial Infection']
cm = confusion_matrix(y_test, y_pred_bi, labels=class_names)
plot_confusion_matrix(cm, classes=class_names)
```

```python
print(classification_report(y_test, y_pred_bi))
```

### Most Important Features

```python
def most_informative_feature_for_class(vectorizer, classifier, n=10):
    # Get the feature names from the vectorizer
    feature_names = vectorizer.get_feature_names_out()

    # Get the feature importances from the classifier
    importances = classifier.named_steps['Random Forest'].feature_importances_

    # Get the top n features based on importance
    topn = sorted(zip(importances, feature_names), reverse=True)[:n]

    # Print the most informative features
    for importance, feat in topn:
        print(f"{feat}: {importance:.4f}")
```

```python
most_informative_feature_for_class(clf.named_steps['vectorizer_tfidf'], clf)
```
